Machine Learning:
- Feature Engineering: créer manuellement;
- Data Processing: fonctionne bien avec de petites quantités de données propres et structurées;
- Scalability: moins adapté aux très grands ensembles de données;
- Pattern Discovery: dépend fortement de l’extraction manuelle des caractéristiques;
- Computational Requirements: CPU suffisant. 

Deep Learning:
- Feature Engineering: apprentissage automatique des caractéristiques pertinentes;
- Data Processing: nécessite de grandes quantités de données, souvent non structurées;
- Scalability: très scalable grâce aux architectures neuronales profondes et aux GPU;
- Pattern Discovery: apprend des motifs complexes et hiérarchiques automatiquement;
- Computational Requirements: nécessite des GPU ou TPU pour un apprentissage efficace. 


ML traditionnel est plus adapté pour: Détection de fraude bancaire
- Les fraudes suivent souvent des motifs répétitifs, et des algorithmes comme les forêts aléatoires ou les modèles linéaires peuvent identifier efficacement les transactions suspectes avec peu de données.

Deep Learning est plus adapté pour: Reconnaissance faciale
- L’identification des visages implique la reconnaissance de motifs complexes et non linéaires dans des images, une tâche pour laquelle les réseaux de neurones convolutifs (CNN) sont bien plus efficaces.

Le Deep Learning excelle dans le traitement des données non structurées, comme les images, le texte et l’audio, car il peut apprendre directement à partir des données brutes sans nécessiter d’ingénierie des caractéristiques. Contrairement aux modèles traditionnels qui demandent une préparation et une sélection manuelles des variables pertinentes, les réseaux neuronaux profonds extraient automatiquement des représentations hiérarchiques des données. De plus, leur capacité à identifier des motifs complexes leur permet de surpasser les méthodes classiques dans des domaines comme la vision par ordinateur et le traitement du langage naturel.

- Couche d’entrée : 3 neurones (correspondant aux caractéristiques en entrée).
- Couche cachée : 4 neurones (traitent les entrées grâce aux poids, biais et fonctions d’activation).
- Couche de sortie : 2 neurones (produisent les prédictions finales).
- Connexions : Chaque neurone est connecté aux neurones de la couche suivante par des poids.
- Biais : Valeurs supplémentaires ajoutées avant d’appliquer la fonction d’activation.
- Fonction d’activation : Introduit de la non-linéarité pour permettre au réseau d’apprendre des modèles complexes.

L’information circule vers l’avant à travers les différentes couches du réseau. D’abord, les valeurs d’entrée sont multipliées par des poids, additionnées à des biais, puis transformées par une fonction d’activation dans la couche cachée. Ces nouvelles valeurs sont transmises à la couche de sortie, qui applique une autre fonction d’activation pour générer des prédictions finales. Ce processus permet au réseau d’apprendre et de modéliser des relations complexes entre les entrées et les sorties.

```python
# Exercise 4
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline

# Générer les données
np.random.seed(0)
x = np.arange(-1, 1, 0.1)
y = -x**2 + np.random.normal(0, 0.05, len(x))

# Séparation en jeu d'entraînement et de test
x_train, x_test = x[:12], x[12:]
y_train, y_test = y[:12], y[12:]

def polynomial_fit(degree):
    """Crée un modèle polynomial de degré spécifié et l'entraîne sur les données d'entraînement."""
    model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
    model.fit(x_train.reshape(-1, 1), y_train)
    return model

def plot_polyfit(degree):
    """Affiche le modèle polynomial ajusté sur les données d'entraînement et de test."""
    model = polynomial_fit(degree)
    x_range = np.linspace(-1, 1, 100).reshape(-1, 1)
    y_pred = model.predict(x_range)
    
    plt.scatter(x_train, y_train, color='blue', label='Train Set')
    plt.scatter(x_test, y_test, color='red', label='Test Set')
    plt.plot(x_range, y_pred, color='green', label=f'Poly Degree {degree}')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.title(f'Polynomial Regression (Degree {degree})')
    plt.show()

# Visualisation pour les degrés 1, 7 et 11
for d in [1, 7, 11]:
    plot_polyfit(d)
```

```python
# Exercise 5
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score, KFold

np.random.seed(0)
x = np.arange(-1, 1.1, 0.1).reshape(-1, 1)
y = -x**2 + np.random.normal(0, 0.05, len(x)).reshape(-1, 1)

# Initialisation des listes pour stocker les erreurs RMSE
cv_rmse = []

# Définition du nombre de folds pour la cross-validation
kf = KFold(n_splits=5, shuffle=True, random_state=0)  # 5-fold cross-validation

# Boucle sur les degrés de polynôme de 1 à 11
for degree in range(1, 12):
    poly = PolynomialFeatures(degree)
    x_poly = poly.fit_transform(x)  # Transformer toutes les données
    
    model = LinearRegression()
    
    # Calcul des scores de cross-validation (négatif car cross_val_score minimise l'erreur)
    scores = cross_val_score(model, x_poly, y, scoring='neg_root_mean_squared_error', cv=kf)
    
    # Stocker la moyenne des erreurs RMSE
    mean_rmse = -scores.mean()
    cv_rmse.append(mean_rmse)

# Tracer les erreurs RMSE en fonction du degré du polynôme
plt.figure(figsize=(8, 5))
plt.plot(range(1, 12), cv_rmse, label='Cross-Validation RMSE', marker='o', color='b')
plt.yscale('log')  # Échelle logarithmique
plt.xlabel('Degré du polynôme')
plt.ylabel('RMSE')
plt.title('Erreur RMSE en fonction du degré du polynôme (Cross-Validation)')
plt.legend()
plt.show()

# Identification du degré avec l'erreur minimale
best_degree = np.argmin(cv_rmse) + 1  # +1 car les indices commencent à 0
print(f"Le meilleur degré avec cross-validation est : {best_degree}")
```
