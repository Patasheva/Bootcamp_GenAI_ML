```python
import pandas as pd 

happiness_world_2019 = pd.read_csv('/Users/patash/PSTB/Week_2/day_4/World_Happiness_Report_2019.csv')
happiness_world_2019.head()
```

```python
happiness_world_2019.info()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 5))
plt.scatter(happiness_world_2019["Social support"], happiness_world_2019["Score"], color="blue", alpha=0.7, edgecolors="black")

plt.xlabel("Social Support")
plt.ylabel("Score")
plt.title("Relationship Between Social Support and Score")
plt.grid(True, linestyle="--", alpha=0.6)

plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].bar(happiness_world_2019["Country or region"], happiness_world_2019["GDP per capita"], color="royalblue")
axes[0].set_title("GDP per Capita by Country")
axes[0].set_ylabel("GDP per Capita ($)")
axes[0].tick_params(axis="x", rotation=30)

axes[1].bar(happiness_world_2019["Country or region"], happiness_world_2019["Healthy life expectancy"], color="seagreen")
axes[1].set_title("Healthy Life Expectancy by Country")
axes[1].set_ylabel("Healthy Life Expectancy (Years)")
axes[1].tick_params(axis="x", rotation=30)

plt.tight_layout()
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(10, 5))

ax1.bar(happiness_world_2019["Country or region"], happiness_world_2019["GDP per capita"], color="royalblue", alpha=0.7, label="GDP per Capita")
ax1.set_ylabel("GDP per Capita ($)", color="royalblue")
ax1.tick_params(axis="y", labelcolor="royalblue")

ax2 = ax1.twinx()
ax2.plot(happiness_world_2019["Country or region"], happiness_world_2019["Healthy life expectancy"], color="seagreen", marker="o", linestyle="-", linewidth=2, markersize=8, label="Healthy Life Expectancy")
ax2.set_ylabel("Healthy Life Expectancy (Years)", color="seagreen")
ax2.tick_params(axis="y", labelcolor="seagreen")


plt.title("Comparison of GDP per Capita and Healthy Life Expectancy by Country")
fig.tight_layout()
plt.show()
```
