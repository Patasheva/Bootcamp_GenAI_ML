```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4]
y = [2, 3, 4, 5]

plt.plot(x,y, color="red")
plt.show
```

```python
plt.plot(x, y, color='green', linestyle='dashed', linewidth=2, marker='o', markerfacecolor='blue', markersize=12)
plt.show()
```

```python
categories = ['Apples', 'Bananas', 'Oranges', 'Grapes']
values = [20, 35, 30, 15]

plt.bar(categories, values)
plt.title('Fruit Sales Data')
plt.xlabel('Fruit')
plt.ylabel('Sales')
plt.show()
```

```python
plt.bar(categories, values, color=['red', 'green', 'blue', 'purple'], edgecolor='black', hatch='/')
plt.title('Customized Fruit Sales Data')
plt.xlabel('Fruit')
plt.ylabel('Sales')
plt.show()
```

```python
import numpy as np

data = np.random.normal(0, 1, 1000)
plt.hist(data, bins=30, edgecolor='black')
plt.title("Histogram of Normally Distributed Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
brands = ['Apple', 'Samsung', 'Huawei', 'Xiaomi', 'Others']
market_share = [40, 30, 15, 10, 5]

plt.pie(market_share, labels=brands, autopct='%1.1f%%')
plt.title('Smartphone Market Share 2023')
plt.show()
```

```python
explode = (0.1, 0, 0, 0, 0)
plt.pie(market_share, labels=brands, autopct='%1.1f%%', explode=explode)
plt.show()
```

```python
import numpy as np

data = np.random.randn(100)
plt.boxplot(data)
plt.title('Basic Box Plot')
plt.ylabel('Values')
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1)
ax1.plot([1, 2, 3], [1, 4, 9])
ax1.set_title('First Subplot')
ax2.plot([1, 2, 3], [1, 2, 3])
ax2.set_title('Second Subplot')
plt.show()
```

```python
x = np.linspace(0, 10, 10) 
y = x**2  

plt.figure(figsize=(8, 5))
plt.plot(x, y, label='$y = x^2$', color='blue', linestyle='-')

plt.xlabel("x")
plt.ylabel("y")
plt.title("Courbe de la fonction y = x^2")

plt.legend()
plt.show()
```

```python
products = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
sales = [150, 220, 180, 90, 250]
colors = ['blue', 'green', 'red', 'purple', 'orange']
patterns = ['/', '\\', '|', '-', '+']

plt.figure(figsize=(8, 5))
bars = plt.bar(products, sales, color=colors)

for bar, pattern in zip(bars, patterns):
    bar.set_hatch(pattern)

plt.xlabel("Produits")
plt.ylabel("Ventes")
plt.title("Ventes des Produits")
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
data = np.random.normal(loc=50, scale=15, size=500) 
plt.figure(figsize=(8, 5))
plt.hist(data, bins=20, color='skyblue', edgecolor='black', alpha=0.7)

plt.xlabel("Valeurs")
plt.ylabel("Fréquence")
plt.title("Histogramme de la distribution normale")
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
activities = ["Travail", "Sommeil", "Loisirs", "Sport", "Autres"]
time_spent = [8, 7, 4, 2, 3] 
colors = ['gold', 'lightblue', 'lightcoral', 'limegreen', 'violet']

plt.figure(figsize=(7, 7))
plt.pie(time_spent, labels=activities, autopct='%1.1f%%', colors=colors, shadow=True, startangle=90)

plt.title("Répartition du Temps Quotidien")
plt.show()
```
