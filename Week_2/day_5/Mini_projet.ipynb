```python
# Which states have the most sales? 
import matplotlib.pyplot as plt
import seaborn as sns

sales_by_state = sales_superstore.groupby('State')['Sales'].sum().sort_values(ascending=False)

# bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x=sales_by_state.index, y=sales_by_state.values, palette="Blues_r")

plt.xticks(rotation=90)
plt.xlabel("State")
plt.ylabel("Total Sales $")
plt.title("Total Sales per State")
plt.show()
```

```python
# What is the difference between New York and California in terms of sales and profit? 

selected_states = sales_superstore[sales_superstore['State'].isin(['California', 'New York'])]
profit_by_state = selected_states.groupby('State')['Profit'].sum()

plt.figure(figsize=(6, 4))
sns.barplot(x=profit_by_state.index, y=profit_by_state.values, palette="coolwarm")

plt.xlabel("State")
plt.ylabel("Total Profit $")
plt.title("Comparison of Total Profit: California vs New York")
plt.show()
```

```python
# Who is an outstanding customer in New York?
ny_customers = sales_superstore[sales_superstore['State'] == 'New York']

customer_sales = ny_customers.groupby(['Customer ID', 'Customer Name']).agg({'Sales': 'sum', 'Profit': 'sum'})

outstanding_customer = customer_sales.sort_values(by='Sales', ascending=False).head(5)

print(outstanding_customer)
```

```python
# Are there any differences among states in profitability?
profit_by_state = sales_superstore.groupby('State')['Profit'].sum().sort_values(ascending=False)

# bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x=profit_by_state.index, y=profit_by_state.values, palette="Blues_r")

plt.xticks(rotation=90)
plt.xlabel("State")
plt.ylabel("Total Profit $")
plt.title("Total Profit per State")
plt.show()
```

```python
# Can we apply Pareto principle to customers and Profit ? 
import numpy as np

profit_per_customer_sorted = sales_superstore.groupby('Customer ID')['Profit'].sum().sort_values(ascending=False)

total_profit = profit_per_customer_sorted.sum()

num_top_customers = int(len(profit_per_customer_sorted) * 0.2)

top_20_profit = profit_per_customer_sorted.iloc[:num_top_customers].sum()

profit_80_percent = total_profit * 0.8

plt.figure(figsize=(8, 5))
plt.bar(["Top 20% Customers", "80% of Total Profit"], [top_20_profit, profit_80_percent], color=['blue', 'red'])

plt.ylabel("Profit ($)")
plt.title("Comparison of Profit: Top 20% Customers vs 80% of Total Profit")
plt.ylim(0, max(top_20_profit, profit_80_percent) * 1.1) 

for i, v in enumerate([top_20_profit, profit_80_percent]):
    plt.text(i, v + (total_profit * 0.02), f"{v:,.2f} $", ha='center', fontsize=12, fontweight='bold')
plt.show()
```

```python
# Top 20 cities by Sales
top_20_cities_sales = sales_superstore.groupby('City')['Sales'].sum().sort_values(ascending=False).head(20)

plt.figure(figsize=(10, 8))
top_20_cities_sales.plot(kind='barh', color='skyblue')
plt.title('Top 20 Cities by Sales')
plt.xlabel('Sales ($)')
plt.ylabel('City')
plt.gca().invert_yaxis() 

for index, value in enumerate(top_20_cities_sales):
    plt.text(value, index, f'{value:.2f}', va='center', ha='left', color='black')
plt.show()
```

```python
# Top 20 cities by Profit
top_20_cities_profit = sales_superstore.groupby('City')['Profit'].sum().sort_values(ascending=False).head(20)

plt.figure(figsize=(10, 8))
top_20_cities_profit.plot(kind='barh', color='lightgreen')
plt.title('Top 20 Cities by Profit')
plt.xlabel('Profit ($)')
plt.ylabel('City')
plt.gca().invert_yaxis()  

for index, value in enumerate(top_20_cities_profit):
    plt.text(value, index, f'{value:.2f}', va='center', ha='left', color='black')
plt.show()
```

```python
# Top 20 customers by Sales
top_20_customers_sales = sales_superstore.groupby('Customer ID')['Sales'].sum().sort_values(ascending=False).head(20)

plt.figure(figsize=(10, 8))
top_20_customers_sales.plot(kind='barh', color='blue')
plt.title('Top 20 Customers by Sales')
plt.xlabel('Sales ($)')
plt.ylabel('Customer ID')
plt.gca().invert_yaxis()  

for index, value in enumerate(top_20_customers_sales):
    plt.text(value, index, f'{value:.2f}', va='center', ha='left', color='black')
plt.show()
```

```python
customer_sales = sales_superstore.groupby('Customer ID')['Sales'].sum().sort_values(ascending=False)

cumulative_sales = customer_sales.cumsum()
cumulative_sales_percentage = cumulative_sales / cumulative_sales.max() 

num_customers = len(customer_sales)
cumulative_customers_percentage = np.arange(1, num_customers + 1) / num_customers 

plt.figure(figsize=(8, 6))
plt.plot(cumulative_customers_percentage * 100, cumulative_sales_percentage * 100, label="Lorenz Curve", color="blue")
plt.plot([0, 100], [0, 100], linestyle="--", color="gray", label="Perfect Equality")

plt.xlabel("Percentage of Customers")
plt.ylabel("Cumulative Percentage of Sales")
plt.title("Lorenz Curve: Distribution of Sales per Customer")
plt.legend()
plt.grid()

plt.show()
```
