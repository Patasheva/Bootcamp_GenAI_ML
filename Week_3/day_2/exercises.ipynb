```python
import numpy as np
import pandas as pd
from scipy.stats import skew, kurtosis

np.random.seed(42) 
data = np.random.uniform(0, 100, 1000)

# 1. Sampling (échantillonnage aléatoire de 20% des données)
sampled_data = np.random.choice(data, size=int(0.2 * len(data)), replace=False)

# 1.1 Shuffling (mélange des données échantillonnées)
np.random.shuffle(sampled_data)

# 2. Calcul des statistiques descriptives
mean_value = np.mean(data)
median_value = np.median(data)
variance_value = np.var(data)
skewness_value = skew(data)
kurtosis_value = kurtosis(data)

# Affichage des résultats
print("Sampled Data (after shuffling):\n", sampled_data[:10])  # Affichage des 10 premiers éléments
print("\nDescriptive Statistics:")
print(f"Mean: {mean_value:.2f}")
print(f"Median: {median_value:.2f}")
print(f"Variance: {variance_value:.2f}")
print(f"Skewness: {skewness_value:.2f}")
print(f"Kurtosis: {kurtosis_value:.2f}")
```

```python
date_range = pd.date_range(start="2022-01-01", end="2022-12-31", freq="D")

np.random.seed(42)
temperature = 10 + 25 * np.sin(np.linspace(0, 2 * np.pi, len(date_range))) + np.random.uniform(-3, 3, len(date_range))

humidity = np.random.uniform(30, 90, len(date_range))

df = pd.DataFrame({"Date": date_range, "Temperature": temperature, "Humidity": humidity})

print(df.head())
```

```python
# Statistiques descriptives
mean_temp = df["Temperature"].mean()
median_temp = df["Temperature"].median()
std_temp = df["Temperature"].std()

mean_hum = df["Humidity"].mean()
median_hum = df["Humidity"].median()
std_hum = df["Humidity"].std()

max_temp_date = df.loc[df["Temperature"].idxmax(), "Date"]
max_hum_date = df.loc[df["Humidity"].idxmax(), "Date"]

print("\nStatistiques:")
print(f"Moyenne Température: {mean_temp:.2f}°C, Médiane: {median_temp:.2f}°C, Écart-type: {std_temp:.2f}")
print(f"Moyenne Humidité: {mean_hum:.2f}%, Médiane: {median_hum:.2f}%, Écart-type: {std_hum:.2f}")
print(f"Jour le plus chaud: {max_temp_date}")
print(f"Jour le plus humide: {max_hum_date}")

plt.figure(figsize=(12, 5))
plt.plot(df["Date"], df["Temperature"], label="Température (°C)", color="red")
plt.plot(df["Date"], df["Humidity"], label="Humidité (%)", color="blue", alpha=0.6)
plt.xlabel("Date")
plt.ylabel("Valeurs")
plt.title("Température et Humidité au cours de l'année 2022")
plt.legend()
plt.show()
```

```python
# Calcul de la moyenne mobile sur 7 jours
df["Temp_MA7"] = df["Temperature"].rolling(window=7).mean()
df["Hum_MA7"] = df["Humidity"].rolling(window=7).mean()

# BONUS : Visualisation des tendances avec la moyenne mobile
plt.figure(figsize=(12, 5))
plt.plot(df["Date"], df["Temperature"], label="Température (Originale)", color="red", alpha=0.5)
plt.plot(df["Date"], df["Temp_MA7"], label="Température (Moyenne mobile 7j)", color="darkred")
plt.plot(df["Date"], df["Humidity"], label="Humidité (Originale)", color="blue", alpha=0.5)
plt.plot(df["Date"], df["Hum_MA7"], label="Humidité (Moyenne mobile 7j)", color="darkblue")
plt.xlabel("Date")
plt.ylabel("Valeurs")
plt.title("Température et Humidité avec moyenne mobile sur 7 jours")
plt.legend()
plt.show()
```

```python
import numpy as np
import scipy.stats as stats
import time

# Création d'un grand tableau de nombres flottants (1 million de valeurs entre 0 et 1)
np.random.seed(42)
large_array = np.random.rand(1_000_000)
print(large_array)
```

```python
# Mesurer le temps d'exécution avec NumPy
start_time = time.time()
mean_numpy = np.mean(large_array)
std_numpy = np.std(large_array)
numpy_time = time.time() - start_time

# Mesurer le temps d'exécution avec SciPy
start_time = time.time()
mean_scipy = stats.tmean(large_array) 
std_scipy = stats.tstd(large_array)   
scipy_time = time.time() - start_time

# Affichage des résultats
print(f"Moyenne avec NumPy: {mean_numpy:.6f}, Écart-type avec NumPy: {std_numpy:.6f}")
print(f"Moyenne avec SciPy: {mean_scipy:.6f}, Écart-type avec SciPy: {std_scipy:.6f}")
print(f"Temps d'exécution NumPy: {numpy_time:.6f} sec")
print(f"Temps d'exécution SciPy: {scipy_time:.6f} sec")
```

```python
import pandas as pd
import numpy as np

# Simulating stock prices for a non-existent stock (DVLPRS)
np.random.seed(0)  

# Generating 100 days of stock prices
dates = pd.date_range(start='2023-01-01', periods=100)
prices = np.random.normal(loc=100, scale=10, size=len(dates))  

stock_data = pd.DataFrame({
    'Date': dates,
    'DVLPRS_Price': prices
})

stock_array = stock_data['DVLPRS_Price'].to_numpy()
print(stock_data.head())
print(stock_array[:5]) 
```

```python

mean_price = np.mean(stock_array)
median_price = np.median(stock_array)
variance_price = np.var(stock_array)
std_dev_price = np.std(stock_array)
skewness_price = skew(stock_array)
kurtosis_price = kurtosis(stock_array)

print(f"Mean Stock Price: {mean_price:.2f}")
print(f"Median Stock Price: {median_price:.2f}")
print(f"Variance of Stock Price: {variance_price:.2f}")
print(f"Standard Deviation of Stock Price: {std_dev_price:.2f}")
print(f"Skewness of Stock Price: {skewness_price:.2f}")
print(f"Kurtosis of Stock Price: {kurtosis_price:.2f}")
```
