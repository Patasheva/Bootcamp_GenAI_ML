```python
import pandas as pd 

df = pd.read_csv('/Users/patash/PSTB/Week_3/day_2/global_power_plant_database.csv')
print(df.head())
```

```python
df.info()
```

```python
# For numerical columns
for col in df.select_dtypes(include=['float64', 'int64']).columns:
    df[col].fillna(df[col].median(), inplace=True)
# For categorical columns
for col in df.select_dtypes(include=['object']).columns:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
df.info()
```

```python
df.describe()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
print(df.columns)
print(df["primary_fuel"].unique())
print(df["country"].unique())
```

```python
country_counts = df["country"].value_counts()

plt.figure(figsize=(12, 6))
sns.barplot(x=country_counts.index[:10], y=country_counts.values[:10], palette="Blues_r")
plt.xticks(rotation=45)
plt.xlabel("Country")
plt.ylabel("Number of Power Plants")
plt.title("Top 10 Countries with the Most Power Plants")
plt.show()
```

```python
fuel_counts = df["primary_fuel"].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(fuel_counts, labels=fuel_counts.index, autopct="%1.1f%%", colors=sns.color_palette("Set2"))
plt.title("Distribution of Power Plants by Fuel Type")
plt.show()
```
