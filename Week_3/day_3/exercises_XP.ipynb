```python
# Exercise 2
from scipy import stats
import numpy as np

data = np.array([12, 15, 13, 12, 18, 20, 22, 21])
mean = np.mean(data)
median = np.median(data)
variance = np.var(data, ddof=1) 
std_dev = np.std(data, ddof=1)  
# Display results
print("Mean:", mean)
print("Median:", median)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
```

```python
# Exercise 3
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mean = 50
std_dev = 10

x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)
y = norm.pdf(x, mean, std_dev)

plt.figure(figsize=(8, 5))
plt.plot(x, y, label=f'Normal Distribution\nMean = {mean}, Std Dev = {std_dev}', color='blue')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Normal Distribution Curve')
plt.legend()
plt.grid()
plt.show()
```

```python
# Exercise 4
from scipy.stats import ttest_ind

data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(60, 10, 100)

t_stat, p_val = ttest_ind(data1, data2)

print("T-statistic =", t_stat, "P-value =", p_val)
```

```python
# Exercise 5

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

house_sizes = np.array([50, 70, 80, 100, 120])  
house_prices = np.array([150000, 200000, 210000, 250000, 280000])

slope, intercept, r_value, p_value, std_err = linregress(house_sizes, house_prices)

print(f"Slope: {slope:.2f}")
print(f"Intercept: {intercept:.2f}")
print(f"Coefficient of determination: {r_value}")
```

```python
predicted_price = slope * 90 + intercept
print(f"Predicted Price for 90m² house: {predicted_price:.2f}")

# The slope represents the increase in house price per additional square meter.
```

```python
# Exercise 6

import numpy as np
from scipy.stats import f_oneway

fertilizer_1 = np.array([5, 6, 7, 6, 5])
fertilizer_2 = np.array([7, 8, 7, 9, 8])
fertilizer_3 = np.array([4, 5, 4, 3, 4])

# ANOVA test
F_value, P_value = f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)

print(f"F-value: {F_value:.2f}")
print(f"P-value: {P_value:.4f}")

# If P-value < 0.05, we reject the null hypothesis (meaning at least one fertilizer significantly affects plant growth differently).
# If P-value > 0.05, we fail to reject the null hypothesis (meaning no significant difference between fertilizers).
```

```python
# Exercise 7
from scipy.stats import binom

# Parameters
n = 10  # Number of coin flips
k = 5   # Number of heads
p = 0.5 # Probability of heads

# Calculate probability of exactly 5 heads
probability = binom.pmf(k, n, p)

print(f"Probability of getting exactly 5 heads in 10 flips: {probability:.4f}")
```

```python
# Exercise 8
import pandas as pd
from scipy.stats import pearsonr, spearmanr

data = pd.DataFrame({
    'age': [23, 25, 30, 35, 40],
    'income': [35000, 40000, 50000, 60000, 70000]
})

# Pearson correlation
pearson_corr, pearson_p = pearsonr(data['age'], data['income'])

# Spearman correlation
spearman_corr, spearman_p = spearmanr(data['age'], data['income'])

print(f"Pearson Correlation: {pearson_corr:.4f}, P-value: {pearson_p:.4f}")
print(f"Spearman Correlation: {spearman_corr:.4f}, P-value: {spearman_p:.4f}")
```
