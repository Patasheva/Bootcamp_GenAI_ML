```python
from sklearn.datasets import load_iris
from scipy.stats import ttest_ind
import pandas as pd

iris = pd.read_excel('/Users/patash/PSTB/Week_2/day_2/iris_export.xlsx')
print(iris.head())
```

```python
# Extract petal lengths for Setosa and Versicolor
setosa_petal_length = iris[iris['Species'] == 'Iris-setosa']['PetalLengthCm']
versicolor_petal_length = iris[iris['Species'] == 'Iris-versicolor']['PetalLengthCm']

# T-test
t_stat, p_val = ttest_ind(setosa_petal_length, versicolor_petal_length)

print("Iris Petal Length Comparison: T-statistic =", t_stat, "P-value =", p_val)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

x = np.array([2, 4, 6, 8, 10])
y = np.array([3, 5, 7, 9, 11])
print(x, y)
```

```python
slope, intercept, r_value, p_value, std_err = linregress(x, y)
print(f"Slope: {slope}")
print(f"Intercept: {intercept}")
print(f"R-squared: {r_value**2}") 
```

```python
y_pred = slope * x + intercept

plt.scatter(x, y, color="blue", label="Data Points")

plt.plot(x, y_pred, color="red", linestyle="--", label="Regression Line")

plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Linear Regression: y = {:.2f}x + {:.2f}".format(slope, intercept))
plt.legend()
plt.show()
```
