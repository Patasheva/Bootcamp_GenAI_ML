```python
print(credit_card.duplicated().sum())
credit_card_clean = credit_card.drop_duplicates()
credit_card_clean.info()
```

```python
credit_card_clean.describe()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6, 4))
ax = sns.countplot(x=credit_card_clean['Class'], palette='Blues')

for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', 
                (p.get_x() + p.get_width() / 2, p.get_height()), 
                ha='center', va='bottom', fontsize=12, fontweight='bold')

plt.xticks([0, 1], ['Non-Fraude (0)', 'Fraude (1)'])
plt.xlabel("Class")
plt.ylabel("Nombre d'occurrences")
plt.title("Distribution de la variable cible")

plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.boxplot(y=credit_card_clean[credit_card_clean['Class'] == 0]['Amount'], ax=axes[0], color='blue')
axes[0].set_title("Transactions Non Frauduleuses")
axes[0].set_ylabel("Montant (€)")

sns.boxplot(y=credit_card_clean[credit_card_clean['Class'] == 1]['Amount'], ax=axes[1], color='red')
axes[1].set_title("Transactions Frauduleuses")
axes[1].set_ylabel("Montant (€)")

plt.tight_layout()  # Ajuste l'affichage
plt.show()
```

```python

# Sélectionner les colonnes V1 à V28
features = [f'V{i}' for i in range(1, 29)]

fig, axes = plt.subplots(nrows=7, ncols=4, figsize=(16, 20))  

for i, feature in enumerate(features):
    row, col = divmod(i, 4)  
    sns.histplot(credit_card_clean[feature], bins=50, kde=True, ax=axes[row, col], color='blue')
    axes[row, col].set_title(f'Distribution de {feature}')

plt.tight_layout()
plt.show()
```

```python
pairs = [('V1', 'V2'), ('V2', 'V3'), ('V3', 'V4'), ('V4', 'V5'),  # etc.
         ('V5', 'V6'), ('V6', 'V7'), ('V7', 'V8'), ('V8', 'V9'),
         ('V9', 'V10'), ('V10', 'V11'), ('V11', 'V12'), ('V12', 'V13'),
         ('V13', 'V14'), ('V14', 'V15'), ('V15', 'V16'), ('V16', 'V17'),
         ('V17', 'V18'), ('V18', 'V19'), ('V19', 'V20'), ('V20', 'V21'),
         ('V21', 'V22'), ('V22', 'V23'), ('V23', 'V24'), ('V24', 'V25'),
         ('V25', 'V26'), ('V26', 'V27'), ('V27', 'V28')]

# Calcul des corrélations pour chaque paire
corr_values = []
for var1, var2 in pairs:
    corr = credit_card_clean[var1].corr(credit_card_clean[var2])
    corr_values.append((var1, var2, corr))

# Affichage des résultats sous forme de graphique
corr_df = pd.DataFrame(corr_values, columns=['Variable 1', 'Variable 2', 'Corrélation'])

# Affichage des corrélations sous forme de barre
plt.figure(figsize=(12, 6))
sns.barplot(x='Corrélation', y='Variable 1', data=corr_df, hue='Variable 2', dodge=False, palette='coolwarm')
plt.title("Corrélations entre paires de variables V1 à V28")
plt.xlabel("Corrélation")
plt.ylabel("Variables")
plt.show()
```
