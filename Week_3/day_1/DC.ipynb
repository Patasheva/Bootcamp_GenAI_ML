```python
import numpy as np
import pandas as pd
```

```python
cities = [f"City_{i+1}" for i in range(10)]
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

temperature_data = np.random.uniform(low=-5, high=35, size=(10, 12))

temperature_df = pd.DataFrame(data=temperature_data, index=cities, columns=months)

print(temperature_df.head())
```

```python
annual_avg_temp = temperature_df.mean(axis=1)

hottest_city = annual_avg_temp.idxmax()
coldest_city = annual_avg_temp.idxmin()

print("Annual Average Temperature for Each City:\n", annual_avg_temp)
print("\nCity with the Highest Average Temperature:", hottest_city)
print("City with the Lowest Average Temperature:", coldest_city)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

sns.heatmap(temperature_df, cmap="coolwarm", annot=True, fmt=".1f")

plt.xlabel("Months")
plt.ylabel("Cities")
plt.title("Monthly Average Temperatures (°C)")

plt.show()
```
